```python
%sql
-- Defining schema for our master table
CREATE TABLE IF NOT EXISTS bank_master_data_bronze (
    id STRING PRIMARY KEY,
    age STRING,
    experience STRING,
    income STRING,
    zip_code STRING,
    family STRING,
    credit_card_average STRING,
    education STRING,
    mortgage STRING,
    has_personal_loan STRING,
    has_securities_account STRING,
    has_cd_account STRING,
    is_online STRING,
    has_credit_card STRING
);
```

**Note:** All columns have the data type **String**. This ensures that no data is dropped during ingestion from the source.

```python
# Reading Data from Bank Loan csv file
df = spark.read.csv(
    "/Volumes/dbacademy_bank/v01/banking/loan-clean.csv",
    header=True,
    inferSchema=True
)
# Creating a Temp View
df.createOrReplaceTempView("bank_master_data")
```

```python
%sql
-- Deleting Records 
DELETE FROM bank_master_data_bronze
```

```python
%sql
-- Insert Records into our newly created table
INSERT INTO bank_master_data_bronze 
SELECT * 
FROM bank_master_data
```
